**Carga y Limpieza de Datos**

```python
# Creamos funciones para cargar datos desde el archivo comprimido JSON
def load_data_json1(archivo):
    data = []
    with gzip.open(archivo, 'rt', encoding='utf-8') as file:
        for line in file:
            data.append(ast.literal_eval(line.strip()))
    return data

def load_data_json2(archivo):
    data = []
    with gzip.open(archivo, 'rt', encoding='utf-8') as file:
        for line in file:
            data.append(json.loads(line))
    return data
```

```python
# Directorio de datos
data_directory = os.getcwd()

# Nombres de los archivos
archivo1 = os.path.join(data_directory, 'users_items.json.gz')
archivo2 = os.path.join(data_directory, 'user_reviews.json.gz')
archivo3 = os.path.join(data_directory, 'steam_games.json.gz')

# Carga de datos
data1 = load_data_json1(archivo1)
df_item = pd.DataFrame(data1)
df_item.info()

data2 = load_data_json1(archivo2)
df_review = pd.DataFrame(data2)
df_review.info()

data3 = load_data_json2(archivo3)
df_games = pd.DataFrame(data3)
df_games.info()
```

```python
# Limpiando Nulos
# Limpiando el DF - df_item
df_item_clean = df_item.dropna()

# Limpiando el DF - df_review
df_review_clean = df_review.dropna()

# Limpiando el DF - df_games
df_games_clean = df_games.dropna()
```

Generación del Sentiment Analysis

```python
# Descargamos el archivo necesario para generar el análisis esperado
import nltk
nltk.download('vader_lexicon')
```

```python
#Importamos la libreria que necesitamos
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Generamos un nueva instancia la cual nos evaluará los sentimientos de las Reviews
analizer = SentimentIntensityAnalyzer()
''' 
Se genera la columna 'polaridad' la cual mediante el método polarity_scores analizará la columna 'reviews'
devolviendonos esté un valor de tipo Float, el cual representa la fuerza del sentimiento, basado en el texto que visualice 
Polaridad < 0 -> Equivale a un sentimiento negativo
Polaridad = 0 -> Equivale a un sentimiento neutro
Polaridad > 0 -> Equivale a un sentimiento positivo
'''
df_reviews_API['polaridad'] = df_reviews_API['reviews'].apply(lambda r: analizer.polarity_scores(r)['compound'])
```

```python
# Generamos una nueva columna llamada 'analisis_sentimiento' la cual mediante el comando 'sentiment' nos devolverá un valor equivalente al resultado esperado
def sentiment(peso):
    if peso < 0:
        return 0
    elif peso > 0:
        return 2
    else:
        return 1
    
df_reviews_API.loc[:,'analisis_sentimiento'] = df_reviews_API['polaridad'].apply(sentiment)
```

```python
# Visualizamos el DF
df_reviews_API
```

```python
# Borramos las columnas que no nos sirvan y visualizamos el resultado
df_reviews_API.drop(columns=['reviews', 'polaridad'], inplace=True)
df_reviews_API
```

```python
#Convertimos los datos de la columna 'user_id' a valores de tipo strings y visualizamos el resultado
df_reviews_API['user_id'] = df_reviews_API['user_id'].str.strip()
df_reviews_API
```
